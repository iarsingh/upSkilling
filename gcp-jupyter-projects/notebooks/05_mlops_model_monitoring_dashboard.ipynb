# MLOps Model Monitoring Dashboard

Goal: analyze model accuracy, latency, request errors, and drift periods from prediction logs.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(Path('../data/model_monitoring_sample.csv'), parse_dates=['event_date'])
df['prediction_correct'] = (df['actual_label'] == df['predicted_label']).astype(int)
df
```

```python
summary = df.groupby(['model_version', 'drift_period'], as_index=False).agg(
    accuracy=('prediction_correct', 'mean'),
    avg_latency_ms=('latency_ms', 'mean'),
    error_rate=('request_error', 'mean'),
    rows=('prediction_correct', 'count')
)
summary
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
summary.plot(x='model_version', y='accuracy', kind='bar', ax=axes[0], title='Accuracy by Model Version')
summary.plot(x='model_version', y='avg_latency_ms', kind='bar', ax=axes[1], title='Average Latency by Model Version')
plt.tight_layout()
```
