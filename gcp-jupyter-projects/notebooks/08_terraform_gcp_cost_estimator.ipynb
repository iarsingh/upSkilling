# Terraform GCP Cost Estimator

Goal: estimate monthly cloud cost from Terraform-style resource planning data.

```python
from pathlib import Path
import pandas as pd

plan = pd.read_csv(Path('../data/terraform_cost_plan.csv'))
plan['estimated_monthly_cost_usd'] = plan['monthly_quantity'] * plan['unit_cost_usd']
plan
```

```python
env_summary = plan.groupby('environment', as_index=False)['estimated_monthly_cost_usd'].sum()
resource_summary = plan.groupby('resource_type', as_index=False)['estimated_monthly_cost_usd'].sum().sort_values('estimated_monthly_cost_usd', ascending=False)
env_summary, resource_summary
```

```python
total = plan['estimated_monthly_cost_usd'].sum()
print(f'Estimated monthly cost: ${total:,.2f}')
print(f'Estimated annual cost: ${total * 12:,.2f}')
```
