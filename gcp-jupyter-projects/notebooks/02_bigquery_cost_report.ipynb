# BigQuery Cost Report Notebook

Goal: practice FinOps-style reporting using a local CSV first, then replace it with a BigQuery query.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

costs = pd.read_csv(Path('../data/costs.csv'), parse_dates=['usage_date'])
costs
```

```python
service_report = (
    costs.groupby(['service', 'environment'], as_index=False)['cost']
    .sum()
    .sort_values('cost', ascending=False)
)
service_report
```

```python
ax = service_report.groupby('service')['cost'].sum().sort_values().plot(kind='barh', title='Cost by GCP Service')
ax.set_xlabel('Cost')
plt.tight_layout()
```

```python
daily_report = costs.groupby('usage_date', as_index=False)['cost'].sum()
daily_report
```
