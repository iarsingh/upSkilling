# Alert Noise Reduction

Goal: group noisy alerts into fewer incident candidates.

```python
from pathlib import Path
import pandas as pd

alerts = pd.read_csv(Path('../data/alerts.csv'), parse_dates=['timestamp'])
alerts['window'] = alerts['timestamp'].dt.floor('5min')
alerts
```

```python
correlated = alerts.groupby(['window', 'namespace', 'service'], as_index=False).agg(
    alert_count=('alert_id', 'count'),
    alert_names=('alert_name', lambda values: ', '.join(sorted(set(values)))),
    resources=('resource', lambda values: ', '.join(sorted(set(values)))),
)
correlated['noise_reduction_ratio'] = correlated['alert_count'] / 1
correlated.sort_values('alert_count', ascending=False)
```

## Interview Talking Points

- Alert correlation reduces duplicate pages.
- Incidents should be grouped by service, time, dependency, and blast radius.
- Correlated alerts should produce one owner and one action plan.
